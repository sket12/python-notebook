## 3°)  Les fonctions

les fonctions sont définies par lemot réservé "def" de cette manière :


```python
#@title Module 16 : Définition des fonctions { run: "auto", vertical-output: true }
#@markdown Il ne se passe rien quand on fait tourner.
#@markdown Simplement la fonction est chargée dans la machine
def f(x):
  return x**2
```

On peut alors remplacer les calculs récurrents par la construction d'une fonction.

Par exemple si on veut conjecturer la limite de la suite récurrence $\ U(0)=1$ et $\ U(n+1)=f(U(n))\ $ pour $\ n\neq 0$ avec $\ f(x)= \sqrt{1+x} $ pour $x\geq-1$

On commence alors à construire la fonction $f$

```python
import math
def f(x):
  return math.sqrt(1+x) 
```

Puis on calcule les 100 premiers termes en les mettant dans une liste.


```python
u=1
Suite=[u]
n=0
while n<=10:
  Suite.append(f(u))
  u=f(u)
  n=n+1
Suite
```

Il semble bien qu'il y ai  convergence  de la suite , supposons qu'elle le soit et que l'on appelle $ "l" $ sa limite .

Cela voudrait dire que : $ \ U(n) \xrightarrow[{n \to \infty}]{} l$ et donc que $ \ U(n+1) \xrightarrow[{n \to \infty}]{} l$

Dont on déduit que $"l"$ est solution de $l= \sqrt{1+l}$

On peut demander à python de résoudre cette équation.

Ce qui donne le module suivant.

```python
import sympy

# Traducteur de réponse au format de lecture traditionnel de l'algèbre
from IPython.display import Math, HTML
display(HTML("<script src='https://cdnjs.cloudflare.com/ajax/libs/mathjax/2.7.3/"
               "latest.js?config=default'></script>"))
# Simplifie l'écrtire
sympy.init_printing()

# Défini les variables 
l = sympy.Symbol('l')

# Là c'est l'éqaution en question
solution=sympy.solve(l-sympy.sqrt(1+l),l) # résolution
solution,solution[0].evalf()# réponse exacte et approchée
```

La réponse semble être la bonne mais ne pas oublier que c'est python qui a résolu l'équation .

Pour résoudre $\ l= \sqrt{1+l} $ sans l'aide de "python" on commence par observer que la solution , si elle existe , doit être postive du fait qu'une racine carré est positive et aussi  que $l\geq-1$ car l'argument d'une racine est positive ou nul.
On en déduit , par intersection que $l\geq0$ donc  que parmi les réponses que l'on va trouver il ne faudra conserver que les solutions positives.

Ensuite la méthode la plus efficace est d'élever cahque terme de l'égalité à la puissance "2" et donc on obtient  $ {l^2}=1+l $ qui est équivalent à $ {l^2}-l-1=0 $ et donc $\ x_{1,2} = \frac{- (-1) \pm \sqrt{5}}{2} \$.

On obtient bien la réponse proposée par python.

On utilise alors ces instructions pour calculer la fréquence d'apparition d'une face sur un dès avec un certain nombre de faces lancé un certain npmbre de fois 

```python
import random
from random import randint # on importe uniquement la fonction dont on a besoin

def freq(faces, nbrelancer):
  result=[randint(1,faces) for i in range(nbrelancer) ]
  nbreface=[]
  for i in range(1,faces+1):
    compt=0
    for j in range(1,nbrelancer):
      if result[j]==i:
        compt=compt+1
    nbreface.append(compt/nbrelancer)
  return nbreface,result
```
